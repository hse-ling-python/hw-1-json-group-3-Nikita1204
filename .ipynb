## Задание №1 

Открываем файл, читаем его построчно и находим количество твитов по количеству строк.

```python
import json


with open('hw_3_twitter.json', encoding='UTF-8') as f:
    text = f.readlines()

l = len(text)
print('Число твитов:', l)
```

## Задание №2

Вычисляем количество удалённых твиттов по ключевому слову "deleted", находим их процент от общего числа твиттов. Заодно создаём новый список твиттов для дальнейшей работы, исключая удалённые.

```python
tweet = []

for i in range(len(text)):
    tweet.append(json.loads(text[i]))

a = 0
tweets = []

for i in range(len(tweet)):
    if 'delete' in tweet[i]:
        a += 1
    else:
        tweets.append(tweet[i])

c = round(a / l * 100)
print('Число удалённых твитов составляет', c, '%', 'от общего числа твитов.', end='')
```

## Задание №3

С помощью Counter считаем количество твиттов на разных языках по тегу 'language'.

```python
from collections import Counter


for i in range(len(tweets)):
    languages.append(tweets[i]['lang'])

counter_lang = Counter(languages)
languages = []

for key in sorted(counter_lang, key=counter_lang.get, reverse=True):
    languages.append(key)

print('Топ-10 самых часто используемых языков:', end=' ')

for i in range(10):
    print(languages[i], end=', ')
```

## Задание № 4

Находим, какие ники пользователей встречаются в тегах двух и более твиттов, считаем их количество посредством Counter.

```python
users = []

for i in range(len(tweets)):
    users.append(tweets[i]['user']['name'])

counter_users = Counter(users)

a = 0
for i in counter_users:
    if counter_users[i] > 1:
        a += 1
        
if a == 0:
    print('Таких пользователей нет.')
else:
    print('Да, таких пользователей ', a, '.', sep='')
```

## Задание №5 

С помощью нескольких индексов проверяем в каждом твитте наличие хэштегов. Если они присутствуют, добавляем в список и затем подсчитываем Counter.

```python
hashtags = []
   
for i in range(len(tweets)):
    if tweets[i]['entities']['hashtags'] != []:
        hashtags.append(tweets[i]['entities']['hashtags'][0]['text'])

counter_tags = Counter(hashtags)
hashtags = []

for key in sorted(counter_tags, key=counter_tags.get, reverse=True):
    hashtags.append(key)

print('Топ-20 самых часто используемых хэштегов:', end=' ')   

for i in range(20):
    print(hashtags[i], end=', ')
```

## Задание №6

Удаляем пунктуацию из каждого твитта, разбиваем твитты на слова по пробелам, добавляем слова из каждого списка в общий список и составляем частотный словарь с помощью Counter.

```python
from string import punctuation

words = []
punct_signs = """.,1234567890<>@&*«…:»;-—'"“”()^%$#!?№/\|=+_"""

for i in range(len(tweets)):
    if tweets[i]['lang'] == 'en':
        a = tweets[i]['text'].lower()
        for j in punct_signs: 
            a = a.replace(j, '')
            w = a.split()
            for u in w:
                for o in punct_signs: 
                    u = u.replace(o, '')
                words.append(u)
        
counter_words = Counter(words)

for key in sorted(counter_words, key=counter_words.get, reverse=True):
    print(key, counter_words[key])
```

## Задание №7

Добавляем в общий словарь пользователей в качестве ключей и количество их подписчиков в качестве значений. Если ключ встречается больше одного раза, в словарь он не добавляется. Составляем топ с помощью Counter.

```python
users = {}

for i in range(len(tweets)):
    user = {tweets[i]['user']['name']: tweets[i]['user']['followers_count']}
    users.update(user)

counter_followers = Counter(users)
users = []

for key in sorted(counter_followers, key=counter_followers.get, reverse=True):
    users.append(key)
    
print('Топ-10 пользователей с наибольшим количеством подписчиков:', end=' ')

for i in range(10):
    print(users[i], end=', ')
```

## Задание №8

В каждом твитте находим строку под тегом 'source', из которой оставляем только то, что находится между треугольными скобками. Полученные кусочки текста добавляем в список и делаем топ с помощью Counter.

```python
import re


for i in range(len(tweets)):
    a = str((tweets[i]['source']))
    source = re.search(r'(<[^>]+>)(.+)(</a>)', a)
    sources.append(source.group(2))
    
counter_sources = Counter(sources)
sources = []

for key in sorted(counter_sources, key=counter_sources.get, reverse=True):
    sources.append(key)

print('Топ-10 источников:', end=' ')

for i in range(10):
    print(sources[i], end=', ')
```
