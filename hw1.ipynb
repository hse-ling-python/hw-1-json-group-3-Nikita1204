## –ó–∞–¥–∞–Ω–∏–µ ‚Ññ1 

–û—Ç–∫—Ä—ã–≤–∞–µ–º —Ñ–∞–π–ª, —á–∏—Ç–∞–µ–º –µ–≥–æ –ø–æ—Å—Ç—Ä–æ—á–Ω–æ –∏ –Ω–∞—Ö–æ–¥–∏–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ç–≤–∏—Ç–æ–≤ –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É —Å—Ç—Ä–æ–∫.

```python
import json


with open('hw_3_twitter.json', encoding='UTF-8') as f:
    text = f.readlines()

l = len(text)
print('–ß–∏—Å–ª–æ —Ç–≤–∏—Ç–æ–≤:', l)
```

## –ó–∞–¥–∞–Ω–∏–µ ‚Ññ2

–í—ã—á–∏—Å–ª—è–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–¥–∞–ª—ë–Ω–Ω—ã—Ö —Ç–≤–∏—Ç—Ç–æ–≤ –ø–æ –∫–ª—é—á–µ–≤–æ–º—É —Å–ª–æ–≤—É "deleted", –Ω–∞—Ö–æ–¥–∏–º –∏—Ö –ø—Ä–æ—Ü–µ–Ω—Ç –æ—Ç –æ–±—â–µ–≥–æ —á–∏—Å–ª–∞ —Ç–≤–∏—Ç—Ç–æ–≤. –ó–∞–æ–¥–Ω–æ —Å–æ–∑–¥–∞—ë–º –Ω–æ–≤—ã–π —Å–ø–∏—Å–æ–∫ —Ç–≤–∏—Ç—Ç–æ–≤ –¥–ª—è –¥–∞–ª—å–Ω–µ–π—à–µ–π —Ä–∞–±–æ—Ç—ã, –∏—Å–∫–ª—é—á–∞—è —É–¥–∞–ª—ë–Ω–Ω—ã–µ.

```python
tweet = []

for i in range(len(text)):
    tweet.append(json.loads(text[i]))

a = 0
tweets = []

for i in range(len(tweet)):
    if 'delete' in tweet[i]:
        a += 1
    else:
        tweets.append(tweet[i])

c = round(a / l * 100)
print('–ß–∏—Å–ª–æ —É–¥–∞–ª—ë–Ω–Ω—ã—Ö —Ç–≤–∏—Ç–æ–≤ —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç', c, '%', '–æ—Ç –æ–±—â–µ–≥–æ —á–∏—Å–ª–∞ —Ç–≤–∏—Ç–æ–≤.', end='')
```

## –ó–∞–¥–∞–Ω–∏–µ ‚Ññ3

–° –ø–æ–º–æ—â—å—é Counter —Å—á–∏—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ç–≤–∏—Ç—Ç–æ–≤ –Ω–∞ —Ä–∞–∑–Ω—ã—Ö —è–∑—ã–∫–∞—Ö –ø–æ —Ç–µ–≥—É 'language'.

```python
from collections import Counter


for i in range(len(tweets)):
    languages.append(tweets[i]['lang'])

counter_lang = Counter(languages)
languages = []

for key in sorted(counter_lang, key=counter_lang.get, reverse=True):
    languages.append(key)

print('–¢–æ–ø-10 —Å–∞–º—ã—Ö —á–∞—Å—Ç–æ –∏—Å–ø–æ–ª—å–∑—É–µ–º—ã—Ö —è–∑—ã–∫–æ–≤:', end=' ')

for i in range(10):
    print(languages[i], end=', ')
```

## –ó–∞–¥–∞–Ω–∏–µ ‚Ññ 4

–ù–∞—Ö–æ–¥–∏–º, –∫–∞–∫–∏–µ –Ω–∏–∫–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤—Å—Ç—Ä–µ—á–∞—é—Ç—Å—è –≤ —Ç–µ–≥–∞—Ö –¥–≤—É—Ö –∏ –±–æ–ª–µ–µ —Ç–≤–∏—Ç—Ç–æ–≤, —Å—á–∏—Ç–∞–µ–º –∏—Ö –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ—Å—Ä–µ–¥—Å—Ç–≤–æ–º Counter.

```python
users = []

for i in range(len(tweets)):
    users.append(tweets[i]['user']['name'])

counter_users = Counter(users)

a = 0
for i in counter_users:
    if counter_users[i] > 1:
        a += 1
        
if a == 0:
    print('–¢–∞–∫–∏—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –Ω–µ—Ç.')
else:
    print('–î–∞, —Ç–∞–∫–∏—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π ', a, '.', sep='')
```

## –ó–∞–¥–∞–Ω–∏–µ ‚Ññ5 

–° –ø–æ–º–æ—â—å—é –Ω–µ—Å–∫–æ–ª—å–∫–∏—Ö –∏–Ω–¥–µ–∫—Å–æ–≤ –ø—Ä–æ–≤–µ—Ä—è–µ–º –≤ –∫–∞–∂–¥–æ–º —Ç–≤–∏—Ç—Ç–µ –Ω–∞–ª–∏—á–∏–µ —Ö—ç—à—Ç–µ–≥–æ–≤. –ï—Å–ª–∏ –æ–Ω–∏ –ø—Ä–∏—Å—É—Ç—Å—Ç–≤—É—é—Ç, –¥–æ–±–∞–≤–ª—è–µ–º –≤ —Å–ø–∏—Å–æ–∫ –∏ –∑–∞—Ç–µ–º –ø–æ–¥—Å—á–∏—Ç—ã–≤–∞–µ–º Counter.

```python
hashtags = []
   
for i in range(len(tweets)):
    if tweets[i]['entities']['hashtags'] != []:
        hashtags.append(tweets[i]['entities']['hashtags'][0]['text'])

counter_tags = Counter(hashtags)
hashtags = []

for key in sorted(counter_tags, key=counter_tags.get, reverse=True):
    hashtags.append(key)

print('–¢–æ–ø-20 —Å–∞–º—ã—Ö —á–∞—Å—Ç–æ –∏—Å–ø–æ–ª—å–∑—É–µ–º—ã—Ö —Ö—ç—à—Ç–µ–≥–æ–≤:', end=' ')   

for i in range(20):
    print(hashtags[i], end=', ')
```

## –ó–∞–¥–∞–Ω–∏–µ ‚Ññ6

–£–¥–∞–ª—è–µ–º –ø—É–Ω–∫—Ç—É–∞—Ü–∏—é –∏–∑ –∫–∞–∂–¥–æ–≥–æ —Ç–≤–∏—Ç—Ç–∞, —Ä–∞–∑–±–∏–≤–∞–µ–º —Ç–≤–∏—Ç—Ç—ã –Ω–∞ —Å–ª–æ–≤–∞ –ø–æ –ø—Ä–æ–±–µ–ª–∞–º, –¥–æ–±–∞–≤–ª—è–µ–º —Å–ª–æ–≤–∞ –∏–∑ –∫–∞–∂–¥–æ–≥–æ —Å–ø–∏—Å–∫–∞ –≤ –æ–±—â–∏–π —Å–ø–∏—Å–æ–∫ –∏ —Å–æ—Å—Ç–∞–≤–ª—è–µ–º —á–∞—Å—Ç–æ—Ç–Ω—ã–π —Å–ª–æ–≤–∞—Ä—å —Å –ø–æ–º–æ—â—å—é Counter.

```python
from string import punctuation

words = []
punct_signs = """.,1234567890<>@&*¬´‚Ä¶:¬ª;-‚Äî'"‚Äú‚Äù()^%$#!?‚Ññ/\|=+_"""

for i in range(len(tweets)):
    if tweets[i]['lang'] == 'en':
        a = tweets[i]['text'].lower()
        for j in punct_signs: 
            a = a.replace(j, '')
            w = a.split()
            for u in w:
                for o in punct_signs: 
                    u = u.replace(o, '')
                words.append(u)
        
counter_words = Counter(words)

for key in sorted(counter_words, key=counter_words.get, reverse=True):
    print(key, counter_words[key])
```

## –ó–∞–¥–∞–Ω–∏–µ ‚Ññ7

–î–æ–±–∞–≤–ª—è–µ–º –≤ –æ–±—â–∏–π —Å–ª–æ–≤–∞—Ä—å –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤ –∫–∞—á–µ—Å—Ç–≤–µ –∫–ª—é—á–µ–π –∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏—Ö –ø–æ–¥–ø–∏—Å—á–∏–∫–æ–≤ –≤ –∫–∞—á–µ—Å—Ç–≤–µ –∑–Ω–∞—á–µ–Ω–∏–π. –ï—Å–ª–∏ –∫–ª—é—á –≤—Å—Ç—Ä–µ—á–∞–µ—Ç—Å—è –±–æ–ª—å—à–µ –æ–¥–Ω–æ–≥–æ —Ä–∞–∑–∞, –≤ —Å–ª–æ–≤–∞—Ä—å –æ–Ω –Ω–µ –¥–æ–±–∞–≤–ª—è–µ—Ç—Å—è. –°–æ—Å—Ç–∞–≤–ª—è–µ–º —Ç–æ–ø —Å –ø–æ–º–æ—â—å—é Counter.

```python
users = {}

for i in range(len(tweets)):
    user = {tweets[i]['user']['name']: tweets[i]['user']['followers_count']}
    users.update(user)

counter_followers = Counter(users)
users = []

for key in sorted(counter_followers, key=counter_followers.get, reverse=True):
    users.append(key)
    
print('–¢–æ–ø-10 –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å –Ω–∞–∏–±–æ–ª—å—à–∏–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –ø–æ–¥–ø–∏—Å—á–∏–∫–æ–≤:', end=' ')

for i in range(10):
    print(users[i], end=', ')
```

## –ó–∞–¥–∞–Ω–∏–µ ‚Ññ8

–í –∫–∞–∂–¥–æ–º —Ç–≤–∏—Ç—Ç–µ –Ω–∞—Ö–æ–¥–∏–º —Å—Ç—Ä–æ–∫—É –ø–æ–¥ —Ç–µ–≥–æ–º 'source', –∏–∑ –∫–æ—Ç–æ—Ä–æ–π –æ—Å—Ç–∞–≤–ª—è–µ–º —Ç–æ–ª—å–∫–æ —Ç–æ, —á—Ç–æ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –º–µ–∂–¥—É —Ç—Ä–µ—É–≥–æ–ª—å–Ω—ã–º–∏ —Å–∫–æ–±–∫–∞–º–∏. –ü–æ–ª—É—á–µ–Ω–Ω—ã–µ –∫—É—Å–æ—á–∫–∏ —Ç–µ–∫—Å—Ç–∞ –¥–æ–±–∞–≤–ª—è–µ–º –≤ —Å–ø–∏—Å–æ–∫ –∏ –¥–µ–ª–∞–µ–º —Ç–æ–ø —Å –ø–æ–º–æ—â—å—é Counter.

```python
import re


for i in range(len(tweets)):
    a = str((tweets[i]['source']))
    source = re.search(r'(<[^>]+>)(.+)(</a>)', a)
    sources.append(source.group(2))
    
counter_sources = Counter(sources)
sources = []

for key in sorted(counter_sources, key=counter_sources.get, reverse=True):
    sources.append(key)

print('–¢–æ–ø-10 –∏—Å—Ç–æ—á–Ω–∏–∫–æ–≤:', end=' ')

for i in range(10):
    print(sources[i], end=', ')
```
